```python
import pandas as pd
import sqlite3
```

```python
#  using sql lite database - show dataframe of picks.db

conn = sqlite3.connect("pickem.db", check_same_thread=False)
cursor = conn.cursor()

users = pd.read_sql_query("SELECT * FROM users", conn)
picks = pd.read_sql_query("SELECT * FROM picks", conn)

print(' ')
print('Players...')
print(users['name'])
print(' ')
print("-----")
print(' ')
print(picks[['username','pick']])
```

```python
# delete table users

cursor.execute("DROP TABLE picks")
```

```python
# stop cursor

cursor.close()
conn.close()
```
